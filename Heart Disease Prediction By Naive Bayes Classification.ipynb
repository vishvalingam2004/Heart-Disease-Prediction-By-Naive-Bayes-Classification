```python
ds['education'].fillna(1.0,inplace=True)
ds['cigsPerDay'].fillna(0.0,inplace=True)
ds['BPMeds'].fillna(0.0,inplace=True)
ds['totChol'].fillna(240.0,inplace=True)
ds['glucose'].fillna(75.0,inplace=True)
ds['heartRate'].fillna(ds['heartRate'].value_counts().idxmax(),inplace=True)
ds['BMI'].fillna(22.19,inplace=True)
```

```python
ds=ds.astype({"education":'int',"cigsPerDay":'int',"BPMeds":'int'})
```

```python
cor=ds.corr()
cor.style.background_gradient(cmap='winter')
```

```python
from sklearn.model_selection import train_test_split
```

```python
x=ds.iloc[:,0:16]
y=ds["TenYearCHD"]
```

```python
xtrain,xtest,ytrain,ytest=train_test_split(x,y,train_size=0.8,random_state=0)
```

```python
xtrain
```

```python
ytrain
```

```python
from sklearn.naive_bayes import GaussianNB
```

```python
mod=GaussianNB()
```

```python
mod.fit(xtrain,ytrain)
```

```python
ypred=mod.predict(xtest)
print(ypred)
```

```python
from sklearn.metrics import confusion_matrix
```

```python
c=confusion_matrix(ytest,ypred)
print("Confusion Matrix is: \n",c)
```

```python
sns.heatmap(c,annot=True,fmt='g',cmap='Purples',xticklabels=['Yes','No'],yticklabels=['Yes','No'])
plt.xlabel('Actual',fontsize=14)
plt.ylabel('Prediction',fontsize=14)
plt.title("Confusion Matrix",fontsize=18)
plt.show()
```

```python
from sklearn.metrics import accuracy_score
from sklearn.metrics import precision_score
from sklearn.metrics import recall_score
from sklearn.metrics import f1_score
```

```python
acc=accuracy_score(ytest,ypred)
print("Accuracy is:",acc*100)
pre=precision_score(ytest,ypred)
print("Precision:",pre*100)
rec=recall_score(ytest,ypred)
print("Recall:",rec*100)
F1=f1_score(ytest,ypred)
print("F1 Score:",F1*100)
```
